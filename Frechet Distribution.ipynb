```python
import pandas as pd
from datetime import datetime
import numpy as np
import hill_estimator 
import matplotlib.pyplot as plt
```

```python
### Hill plot --- Estimate thresholds
data = pd.read_excel('Data.xlsx', sheet_name='Logreturn Data', 
                     index_col=0, parse_dates=True)

u = np.linspace(0, -0.02222222, 200)

CL = data.columns.tolist()

lst = []
for i in range(len(u)):
    lst.append(data[(data[CL[0]] < u[i])])

gamma = []
m = []
for i in range(len(lst)):
    x = hill_estimator.hill_estimator(lst[i][CL[0]].sort_values(ascending=False)*(-1))
    gamma.append(x[3])
    m.append(x[2])

u_lst = u.tolist()

fig, ax = plt.subplots()
plt.plot(u_lst, gamma)
plt.xlabel('Threshold u')
plt.ylabel('Hill Estimates')
plt.title('Hill Plot for SPX')
```

```python
### summary for threshold

u = [-0.0022, -0.0033, -0.006, -0.0075, -0.0075, -0.008, -0.0055, -0.0025]

loss_tail = data[(data[CL[0]]<-0.0022) & (data[CL[1]]<-0.0033) & (data[CL[2]]<-0.006) &
            (data[CL[3]]<-0.0075) & (data[CL[4]]<-0.0075) & (data[CL[5]]<-0.008) &
            (data[CL[6]]<-0.0055) & (data[CL[7]]<-0.0025)]
print(loss_tail)
print(data)
```

```python
### zt
u = []
for i in range(8):
    u.append(np.mean(data.iloc[:, i]) - np.std(data.iloc[:, i])*0.5)

zt = pd.DataFrame(columns=loss_tail.columns.tolist(), index=loss_tail.index.tolist())
cl = zt.columns.tolist()
for i in range(len(cl)):
    zt[cl[i]] = np.abs(loss_tail.loc[:, cl[i]] - u[i])

print(zt)
```

```python
### parameter gamma
gamma = []
m = []
for i in range(len(cl)):
    x = hill_estimator.hill_estimator(loss_tail[cl[i]].sort_values(ascending=False)*(-1))
    gamma.append(x[3])
    m.append(x[2])

print(gamma)
```

```python
from scipy.stats import invweibull
frechet_cdf = pd.DataFrame(columns=loss_tail.columns.tolist(), index=loss_tail.index.tolist())

for i in range(len(cl)):
    frechet_cdf[cl[i]] = invweibull.cdf(zt[cl[i]], gamma[i]*(-1))

print(frechet_cdf)
```
